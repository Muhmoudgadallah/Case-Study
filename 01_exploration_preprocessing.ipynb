```python
# YearsCode , YearsCodePro

#Nan values 

#String values 

def make_numeric(i) : 
    try : 
        return float(i)
    except : 
        
        if i == 'Less than 1 year' :
            return 0.5
        elif i=='More than 50 years' : 
            return np.random.randint(50,60)
        else : 
            return np.nan

                

df['YearsCode'] = df['YearsCode'].apply(make_numeric)
df['YearsCodePro'] = df['YearsCodePro'].apply(make_numeric)
```

### There is some columns with multipleanswers separated with semicolumns , which need to be separated and refornatted to be easy to deal with . 

## columns with multiple answers : 
* LearnCode 
* DevType
* LanguageHaveWorkedWith
* DatabaseHaveWorkedWith
* PlatformHaveWorkedWith
* WebframeHaveWorkedWith 
* MiscTechHaveWorkedWith
* MiscTechWantToWorkWith
* ToolsTechHaveWorkedWith
* NEWCollabToolsHaveWorkedWith

### in these columns i will make each one of them in separate dataset to be easy to deal with and to avoid making the dataset very large .


```python
def split_answers(col) : 
    
    #split the answers to get list of the different answers
    splitted_col = df[col].str.split(";")
    
    # esvplode : to make each answer in repeated raws with the same Id 
    exploded_col = splitted_col.explode()
    
    # make one Hot encoder to : 
    ''' 1- make a column for eacj single answer 
        2- help to make groupby to find counts,sums , etc..
        3- make it in separate dataset to use it in Analysis 
        4- use this dataset in our modeling '''
    dummies = pd.get_dummies(exploded_col)
    
    #grouping by index to restore the oreginal raws and to find the answers separately 
    new_col = dummies.groupby(level=0).sum()
    return new_col

multiple_answers = ['LearnCode','DevType','LanguageHaveWorkedWith','LanguageWantToWorkWith', 'DatabaseHaveWorkedWith',
                       'PlatformHaveWorkedWith', 'WebframeHaveWorkedWith','MiscTechHaveWorkedWith', 'MiscTechWantToWorkWith',
                       'ToolsTechHaveWorkedWith', 'NEWCollabToolsHaveWorkedWith']

```

```python
#this dict for recording frequency of each category classes 
unique_answers = {}

#loop over the multiple answers to make sub datasets for each column 
for col in multiple_answers : 
    #call split_answers func to get thegrouped dummies 
    new = split_answers(col)
    
    # save the grouped dummies in sub data set in pkl file 
    new.to_pickle(f"..//Datasets//{col}.pkl")
    
    # sum over raws to find the freqency of each category and restore it 
    unique_answers[f"{col}"] = new.sum(axis=0)
    
```

```python
for i in unique_answers : 
    print(i)
    print("-"*20)
    print(unique_answers[i].sort_values(ascending=False))
    print("_"*100)
```

```python
# Save unique_answers in one file to use it latter in Analytics section 
import pickle
with open ("..//Datasets//unique_answers.pkl",'wb') as file:
    pickle.dump(unique_answers,file)
```
